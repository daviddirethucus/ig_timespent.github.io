### - rearrange columns

```python
df.rename(columns={"Instagram name":"name","likes":"a","followings":"b","How often do you post the Instagram Story":"c","How often do you see the Instagram Story from other users?":"d","How often do you glance through the Instagram Shop?":"e","How often do you chat with others on Instagram?":"f","How often do you glance through the 'Explore' or 'Reels' on Instagram?":"g","How often do you spend time on Instagram Advertisement posts?":"h","What is your average daily time spent on Instagram?":"time_spent"},inplace=True)
df.drop(columns=['posts','followers'],inplace=True)
oofs = list(df.columns)
df = df[oofs[0:2]+oofs[-1:]+oofs[-2:-1]+oofs[3:9]+oofs[2:3]]
df.head(1)
```

### - transform the time_spent
Owing to the different format and chaos of the time_spent feature (e.g. 1h26m6s, 5minutes), this column should have a further clean.

```python
time_list = []

for i in df['time_spent']:
    hrs=0
    mins=0

    if i.find('h')!=-1:
        hrs = i[i.find('h')-1] 
    elif i.find('H')!=-1:
        hrs = i[i.find('H')-1]
    else:
        hrs = None
    
    if i.find('m')!=-1:
        if i[i.find('m')-2].isdigit():
            mins = i[i.find('m')-2:i.find('m')]
        else:
            mins = i[i.find('m')-1]
    elif i.find('M')!=-1:
        if i[i.find('M')-2].isdigit():
            mins = i[i.find('M')-2:i.find('M')]
        else:
            mins = i[i.find('M')-1]
    else:
        mins = None
        
    time_list.append([hrs,mins])
    
i=0
for time in time_list:
    if time[0]!=None and time[1]!=None:
        time = str( int(time[0]) * 60 + int(time[1]) )
    elif time[1]==None:
        time = str( int(time[0]) * 60 )
    else:
        time = time[1]
    
    df['time_spent'][i] = time
    i+=1

df.head()
```

```python
x = df.iloc[:,:-1]

x_norm = StandardScaler().fit_transform(x[['a','b']])
x.iloc[:,:2]=x_norm

y = df[['time_spent']] 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=42)

model = XGBRegressor()
model.fit(x_train,y_train)
# cv = RepeatedKFold(n_splits=10,n_repeats=3, random_state=1)
# scores = cross_val_score(model,x_train,y_train,scoring='neg_mean_absolute_error',cv=cv,n_jobs=-1)
```

```python
pred = model.predict(x_test)
print(mean_absolute_error(pred, y_test))
r2 = r2_score(pred,y_test)
print(r2)
```

```python
adj_r2 = 1-(1-r2)*(len(x_train)-1)/(len(x_train)-len(x_train.columns)-1)
adj_r2
```

```python
from sklearn.inspection import permutation_importance
per_im = permutation_importance(model,x_test,y_test,n_repeats=10,random_state=42)
sorted_idx = per_im.importances_mean.argsort()
plt.boxplot(
    per_im.importances[sorted_idx].T,
    vert=False,
    labels=np.array(df.columns)[sorted_idx] )
plt.show()
```
